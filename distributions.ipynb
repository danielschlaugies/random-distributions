# Plots of distributions functions of random distributions

```python
import numpy as np
import numpy.typing as npt
from scipy.special import comb, factorial
import matplotlib.pyplot as plt
```

## Bernoulli distribution

```python
def bernoulli(k: npt.NDArray[np.int_], p: float) -> npt.NDArray[np.float_]:
    return np.where(k == 1, p, 0) + np.where(k == 0, 1 - p, 0)

kk = np.array([0, 1])
pp = bernoulli(kk, 1/3)

plt.scatter(kk, pp)
```

## Geometric distribution

```python
def geometric(k: npt.NDArray[np.int_], p: float) -> npt.NDArray[np.float_]:
    return (1 - p)**(k - 1) * p

kk = np.arange(1, 21)
pp = geometric(kk, 1/3)

plt.scatter(kk, pp)
```

## Binomial distribution

```python
def binomial(k: npt.NDArray[np.int_], n: int, p: float) -> npt.NDArray[np.float_]:
    return comb(n, k) * p ** k * (1 - p)**(n - k)

kk = np.arange(1, 21)
pp = binomial(kk, 10, 2/3)

plt.scatter(kk, pp)
```

## Poisson distribution

```python
def poisson(k: npt.NDArray[np.int_], l: float) -> npt.NDArray[np.float_]:
    return np.exp(-l) * l**k / factorial(k)

kk = np.arange(0, 31)
pp = poisson(kk, 10)

plt.scatter(kk, pp)
```

## Normal distribution

```python
def normal(x: npt.NDArray[np.float_], mu: np.float_, sigma: np.float_) -> npt.NDArray[np.float_]:
    return np.exp(-0.5 * ((x - mu) / sigma)**2) / (sigma * np.sqrt(2 * np.pi))

xx = np.linspace(-4, 10)
yy = normal(xx, 3, 2)

plt.plot(xx, yy)
```
